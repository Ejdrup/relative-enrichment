```python
import numpy as np
import matplotlib.pyplot as plt
import RE_function as RE
import vesicle_function as vf
from scipy import stats
```

## Simulation plots

### Single vesicle example (Fig. 4A)

```python
# Simulate data
vesicle = vf.generate_spherical(100, radius = 30) # Vesicle data
AZ = vf.generate_circle(50, radius = 30) # AZ data
cargo = vf.generate_vesicle_content(50, radius = 20) # Luminal data

# Plot single docked vesicle
fig = plt.figure(dpi=100, figsize=(3, 3))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(vesicle[0,:], vesicle[1,:], vesicle[2,:], s=20, c='dimgrey')
ax.scatter(AZ[0,:], AZ[1,:], AZ[2,:], s=20, c='teal')
ax.scatter(cargo[0,:], cargo[1,:], cargo[2,:], s=20, c='red')
ax.set_xlim(-100,100)
ax.set_ylim(-100,100)
ax.set_zlim(-100,100)
ax.set_xticks([-100,-50,0,50,100])
ax.set_yticks([-100,-50,0,50,100])
ax.set_zticks([-100,-50,0,50,100])
ax.set_zticklabels([])
ax.set_yticklabels([])
ax.set_xticklabels([])
ax.set_xlabel("x", labelpad = -10)
ax.set_ylabel("y", labelpad = -10)
ax.set_zlabel("z", labelpad = -10)
plt.title("Docked vesicle example", fontsize = 10)
plt.legend(("Vesicle","AZ","Luminal"),frameon = True, loc="upper left", prop={'size': 10})
ax.azim = 110
ax.elev = 10
```

### Simulation of 20 vesicles (Fig. 4B, C)

```python
# Simualate data
vesicle_coords, cargo_coords, AZ_coords = vf.vesicle_simulation(1)

# 3D plot
fig = plt.figure(dpi=100, figsize=(3,3))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(AZ_coords[0,:], AZ_coords[1,:], AZ_coords[2,:], marker  = 'o', s=4, lw = 0, c='teal', alpha = 1)
ax.scatter(cargo_coords[0,:], cargo_coords[1,:], cargo_coords[2,:], marker  = 'o', s=4, lw = 0, c='red', alpha = 1)
ax.scatter(vesicle_coords[0,:], vesicle_coords[1,:], vesicle_coords[2,:], marker  = 'o', s=4, lw = 0, c='dimgrey', alpha = 0.5)

fig.text(0.55,0.85,"Vesicle pool, 3D", fontsize = 10, ha = "center")
ax.set_zlim(-200,200)
ax.set_ylim(-400,400)
ax.set_xlim(-400,400)
ax.set_xticks([-400,-200,0,200,400])
ax.set_xticklabels([])
ax.set_yticks([-400,-200,0,200,400])
ax.set_yticklabels([])
ax.set_zticks([-200,0,200])
ax.set_zticklabels([])
ax.set_xlabel("x", labelpad = -10)
ax.set_ylabel("y", labelpad = -10)
ax.set_zlabel("z", labelpad = -10)
ax.azim = 110
ax.elev = 10

plt.tight_layout()

# 2D Projection plot
fig = plt.figure(dpi = 100, figsize=(3,3))
ax = fig.add_subplot(111)

ax.scatter(vesicle_coords[0,:], vesicle_coords[1,:], s=1, c='dimgrey')
ax.scatter(AZ_coords[0,:], AZ_coords[1,:], s=1, c='teal')
ax.scatter(cargo_coords[0,:], cargo_coords[1,:], s=1, c='red')
ax.set_ylim(-400,400)
ax.set_xlim(-400,400)
ax.set_xticks([])
ax.set_yticks([])
ax.set_title("2D projection", fontsize = 10)


plt.tight_layout()
```

### Vesicle localizations color-coded by RE (Fig. 4F & Fig. S3A)

```python
# Calculation of 3D RE

# For AZ
n_points_AZ_3d, area_AZ_3d, dist_AZ_3d,bool_index_AZ = RE.RE3D(vesicle_coords.T, AZ_coords.T)
n_exp_AZ = n_points_AZ_3d/(AZ_coords.shape[1]*area_AZ_3d/800**3)
n_exp_AZ[np.isinf(n_exp_AZ)] = 0.1
n_exp_AZ[n_exp_AZ == 0] = 0.1


# For cargo
n_points_cargo_3d, area_cargo_3d, dist_AZ_3d,bool_index_cargo = RE.RE3D(vesicle_coords.T, cargo_coords.T)
n_exp_cargo = n_points_cargo_3d/(cargo_coords.shape[1]*area_cargo_3d/800**3)
n_exp_cargo[np.isinf(n_exp_cargo)] = 0.1
n_exp_cargo[n_exp_cargo == 0] = 0.1
```

#### Plot in 3D

```python
fig = plt.figure(dpi=100, figsize=(3, 3))
fig.text(0.6,0.88,"Vesicles: color by RE", fontsize = 10, ha = "center")
fig.text(0.2,0.6,"AZ", fontsize = 10, ha = "center", rotation = 90)
fig.text(0.2,0.23,"Luminal", fontsize = 10, ha = "center", rotation = 90)

ax = fig.add_subplot(2,1,1, projection='3d')
sc = ax.scatter(vesicle_coords[0,:][bool_index_AZ], vesicle_coords[1,:][bool_index_AZ], vesicle_coords[2,:][bool_index_AZ],
           c = n_exp_AZ, marker  = 'o', s=3, lw = 0, alpha = 1,cmap = "viridis", vmin = 0, vmax = 10)
ax.set_box_aspect(aspect = (1,1,0.5))
ax.set_zlim(-200,200)
ax.set_ylim(-400,400)
ax.set_xlim(-400,400)
ax.set_xlabel("x", labelpad = -12)
ax.set_ylabel("y", labelpad = -12)
ax.set_zlabel("z", labelpad = -14)
ax.azim = 110
ax.elev = 10

cb_ax = fig.add_axes([0.85, 0.2, 0.02, 0.6])
cbar = fig.colorbar(sc, cax=cb_ax)
cbar.set_ticks([0,5,10])
cbar.set_label("Relative enrichment", rotation=90, labelpad = 0, fontsize = 10)

ax2 = fig.add_subplot(2,1,2, projection='3d')
ax2.scatter(vesicle_coords[0,:][bool_index_cargo], vesicle_coords[1,:][bool_index_cargo], vesicle_coords[2,:][bool_index_cargo],
           c = n_exp_cargo, marker  = 'o', s=3, lw = 0, alpha = 1,cmap = "viridis", vmin = 0, vmax = 10)
ax2.set_box_aspect(aspect = (1,1,0.5))
ax2.set_zlim(-200,200)
ax2.set_ylim(-400,400)
ax2.set_xlim(-400,400)
ax2.set_xticks([-400,-200,0,200,400])
ax2.set_yticks([-400,-200,0,200,400])
ax2.set_zticks([-200,0,200])
ax2.set_xlabel("x", labelpad = -12)
ax2.set_ylabel("y", labelpad = -12)
ax2.set_zlabel("z", labelpad = -14)
ax2.azim = 110
ax2.elev = 10
ax2.patch.set_visible(False)

plt.tight_layout(w_pad = 0, h_pad = -4)

ax.set_zticklabels([])
ax.set_xticklabels([])
ax.set_yticklabels([])

ax2.set_zticklabels([])
ax2.set_xticklabels([])
ax2.set_yticklabels([])
```

## Calculating relative enrichment (RE)

### Simulation of data

```python
# Numer of simulations
no_runs = 5

# Set bin values
max_range = 4
step_size = 0.2
x_range = np.linspace(0,max_range,int(max_range/step_size)+1)

# Pre-allocate space for data (runs x steps x dimension (2d, 3x) x primary species (AZ, cargo))
gather_RE = np.zeros((no_runs,int(max_range/step_size)+1,2,2))

gather_bin = np.zeros((no_runs,int(max_range/step_size)+1,2,2))

# Run simulations
for i in range(no_runs):
    print("Run number: " + str(i+1))
    # Simulate vesicles with cargo and AZ
    vesicle_coords, cargo_coords, AZ_coords = vf.vesicle_simulation(1)
    
    # Calculate RE for each simulation
    n_points_AZ_3d, area_AZ_3d, dist_AZ_3d, _ = RE.RE3D(vesicle_coords.T, AZ_coords.T)
    n_points_cargo_3d, area_cargo_3d, dist_cargo_3d, _ = RE.RE3D(vesicle_coords.T, cargo_coords.T)
    
    n_points_AZ_2d, area_AZ_2d, dist_AZ_2d, _ = RE.RE(vesicle_coords[0:2,:].T, AZ_coords[0:2,:].T)
    n_points_cargo_2d, area_cargo_2d, dist_cargo_2d, _ = RE.RE(vesicle_coords[0:2,:].T, cargo_coords[0:2,:].T)
    
    
    # Bin RE by distance to nearest neighbor
    RE_AZ_3d, bin_AZ_3d, loc_AZ_3d, area_AZ_3d = RE.bin_RE(np.array(n_points_AZ_3d), np.array(area_AZ_3d), np.array(dist_AZ_3d),
                                       max_range,step_size)
    RE_cargo_3d, bin_cargo_3d, loc_cargo_3d, area_cargo_3d = RE.bin_RE(np.array(n_points_cargo_3d), np.array(area_cargo_3d), np.array(dist_cargo_3d),
                                       max_range,step_size)
    
    RE_AZ_2d, bin_AZ_2d, loc_AZ_2d, area_AZ_2d = RE.bin_RE(np.array(n_points_AZ_2d), np.array(area_AZ_2d), np.array(dist_AZ_2d),
                                       max_range,step_size)
    RE_cargo_2d, bin_cargo_2d, loc_cargo_2d, area_cargo_2d = RE.bin_RE(np.array(n_points_cargo_2d), np.array(area_cargo_2d), np.array(dist_cargo_2d),
                                       max_range,step_size)
    
    # Gather all the data for each run
    gather_RE[i,:,0,0] = RE_AZ_2d.ravel()
    gather_RE[i,:,1,0] = RE_AZ_3d.ravel()
    gather_RE[i,:,0,1] = RE_cargo_2d.ravel()
    gather_RE[i,:,1,1] = RE_cargo_3d.ravel()
    
    gather_bin[i,:,0,0] = bin_AZ_2d.ravel()
    gather_bin[i,:,1,0] = bin_AZ_3d.ravel()
    gather_bin[i,:,0,1] = bin_cargo_2d.ravel()
    gather_bin[i,:,1,1] = bin_cargo_3d.ravel()
```

```python
# Calculate mean and standard error of the mean for plots
mean_RE_AZ_2d = np.nanmean(gather_RE[:,:,0,0],axis=0)
sem_RE_AZ_2d = stats.sem(gather_RE[:,:,0,0],axis=0,nan_policy = "omit")
mean_RE_cargo_2d = np.nanmean(gather_RE[:,:,0,1],axis=0)
sem_RE_cargo_2d = stats.sem(gather_RE[:,:,0,1],axis=0,nan_policy = "omit")
mean_RE_AZ_3d = np.nanmean(gather_RE[:,:,1,0],axis=0)
sem_RE_AZ_3d = stats.sem(gather_RE[:,:,1,0],axis=0,nan_policy = "omit")
mean_RE_cargo_3d = np.nanmean(gather_RE[:,:,1,1],axis=0)
sem_RE_cargo_3d = stats.sem(gather_RE[:,:,1,1],axis=0,nan_policy = "omit")
```

### RE plots (Fig. 4D, E)

```python
# RE for 2D projection data
fig, ax1 = plt.subplots(figsize=(3, 3), dpi=100) # subplots rather than plt will be used to fine-tune the output
ax1.plot(x_range,mean_RE_AZ_2d, color = 'teal', label = "AZ")
ax1.plot(x_range,mean_RE_cargo_2d, color = 'red', label = "Luminal")
ax1.fill_between(x_range,mean_RE_AZ_2d-sem_RE_AZ_2d, mean_RE_AZ_2d+sem_RE_AZ_2d, 
                 color = 'teal', alpha=0.4, label='_Hidden label')
ax1.fill_between(x_range,mean_RE_cargo_2d-sem_RE_cargo_2d, mean_RE_cargo_2d+sem_RE_cargo_2d, 
                 color = 'red', alpha=0.4,label='_Hidden label')
ax1.set_zorder(2)
ax1.patch.set_visible(False)
ax1.spines['top'].set_visible(False)
ax1.set_ylabel("Relative Enrichment")
ax1.set_xlim(-0.1, 2.5)
ax1.set_xlabel("NND (nm)")
ax1.legend(frameon=False, prop={'size': 9})
ax1.set_xticks([0,1,2,3])
ax1.set_xticklabels(["1","10","100","1000"])

ax2 = ax1.twinx()
ax2.bar(x_range,np.mean(gather_bin[:,:,0,0],axis=0)/sum(np.mean(gather_bin[:,:,0,0],axis=0)), width = 0.18, align = "center", color = "dimgrey")
ax2.set_zorder(1)
ax2.spines['top'].set_visible(False)
ax2.set_ylabel("Norm. vesicle density")
ax2.set_ylim(0,0.5)
ax1.spines['right'].set_color('dimgrey')
ax2.spines['right'].set_color('dimgrey')
ax2.tick_params(axis='y', colors='dimgrey')
ax2.yaxis.label.set_color('dimgrey')
ax2.set_xlim(-0.1, 2.5)
plt.title("Analysis of 2D projection", fontsize = 10)

plt.tight_layout()
```

```python
# RE for 3D data
fig, ax1 = plt.subplots(figsize=(3, 3), dpi=100) # subplots rather than plt will be used to fine-tune the output
ax1.plot(x_range,mean_RE_AZ_3d, color = 'teal', label = "AZ")
ax1.plot(x_range,mean_RE_cargo_3d, color = 'red', label = "Luminal")
ax1.fill_between(x_range,mean_RE_AZ_3d-sem_RE_AZ_3d, mean_RE_AZ_3d+sem_RE_AZ_3d, 
                 color = 'teal', alpha=0.4, label='_Hidden label')
ax1.fill_between(x_range,mean_RE_cargo_3d-sem_RE_cargo_3d, mean_RE_cargo_3d+sem_RE_cargo_3d, 
                 color = 'red', alpha=0.4,label='_Hidden label')
ax1.set_zorder(2)
ax1.patch.set_visible(False)
ax1.spines['top'].set_visible(False)
ax1.set_ylabel("Relative Enrichment")
ax1.set_xlabel("NND (nm)")
ax1.legend(frameon=False, prop={'size': 9})
ax1.set_xticks([0,1,2,3])
ax1.set_xticklabels(["1","10","100","1000"])

ax2 = ax1.twinx()
ax2.bar(x_range,np.mean(gather_bin[:,:,1,0],axis=0)/sum(np.mean(gather_bin[:,:,1,0],axis=0)), width = 0.18, align = "center", color = "dimgrey")
ax2.set_zorder(1)
ax2.spines['top'].set_visible(False)
ax2.set_ylabel("Norm. vesicle density")
ax2.set_ylim(0,0.5)
ax1.spines['right'].set_color('dimgrey')
ax2.spines['right'].set_color('dimgrey')
ax2.tick_params(axis='y', colors='dimgrey')
ax2.yaxis.label.set_color('dimgrey')
ax2.set_xlim(0.5, 3)
plt.title("Analysis of 3D projection", fontsize = 10)
plt.tight_layout()
```
